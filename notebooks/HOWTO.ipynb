# yfinance HOWTO

### How to effectively use yfinance

## Load Historical Prices

```python
# Loading Yahoo Prices
import yfinance as yf

symbols = ["AAPL", "GOOG", "NVDA"]
start_date = '2022-01-01'
end_date = '2022-12-31'
prices = yf.download(symbols, start=start_date, end=end_date, 
                     actions=True, # Get dividends and splits
                     auto_adjust=False, # Auto-adjust close price for dividends
                     progress=False # show progress bar
                    )

prices
```

```python
# Extract close price for all symbols
close_prices = prices["Close"]
close_prices
```

```python
# Extract all prices for a single symbol
prices_goog = prices.xs("GOOG", level=1, axis=1)
prices_goog
```

## Resample Prices

```python
prices.resample("ME").last()
```

## Compute Unadjusted prices before Splits

```python
# Extract relevant data
close_prices = prices_goog["Close"]  # Already adjusted for splits
splits = prices_goog["Stock Splits"]  # Stock split data

splits = splits.replace(0, 1)  # Ensure no division by zero

# Compute cumulative product of splits to reconstruct raw prices
cumulative_splits = splits[::-1].cumprod()[::-1] 
cumulative_splits
```

```python
# Reconstruct fully unadjusted close prices
unadjusted_close = close_prices * cumulative_splits  # Reverse split adjustments
unadjusted_close
```

## Compute Portfolio Returns

```python
etf_portfolio = {
    "VOO": 30,  # U.S. Stocks (Large-Cap)
    "VB": 10,   # U.S. Stocks (Small-Cap)
    "VEA": 15,  # International Stocks (Developed Markets)
    "VWO": 10,  # Emerging Markets Stocks
    "BND": 15,  # Bonds (U.S. Treasuries)
    "VNQ": 10,  # Real Estate (REITs)
    "GLD": 5,   # Commodities (Gold)
    "VTIP": 5   # Cash or Inflation Protection (TIPS)
}

symbols = list(etf_portfolio.keys())
from_date='2016-01-01'
to_date = '2024-12-31'
prices = yf.download(symbols, start=from_date, end=to_date, auto_adjust=True, progress=False)
# Extract close price for all symbols
close_prices = prices["Close"]
close_prices
```

```python
# Compute returns for each asset
returns = close_prices.pct_change().dropna()
returns
```

```python
# Compute portfolio returns
import numpy as np

weights = np.array(list(etf_portfolio.values())) / 100  # Convert % to decimal

# Compute portfolio returns (dot product of weights and returns)
portfolio_returns = returns.dot(weights)

# Convert to DataFrame
returns["Portfolio Return"] = portfolio_returns

# Display first few rows
returns
```
